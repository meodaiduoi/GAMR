```python


import sqlite3
import pandas as pd

def query_link_state(path):
    con = sqlite3.connect(path)
    cur = con.cursor()
    cur.execute('SELECT * FROM linkcost')
    
    rows = cur.fetchall()
    return rows
    con.close()
    
def query_link_last_state(path):
    con = sqlite3.connect(path)
    cur = con.cursor()
    cur.execute('SELECT * FROM linkcost WHERE time_id = (SELECT MAX(time_id) FROM linkcost)')
    
    rows = cur.fetchall()
    return rows
    con.close()


row = query_link_state('./db/linkcost-12-16-16--17-09-2023.db')
#HostwayInternational

df = pd.DataFrame(row, columns=['timeid', 'src', 'dst', 'delay', 'packetloss', 'link_usage', 'free_bandwidth'])
savedfilename1 = input("ten thuat toan")
savedfilename2 = input("số switch")
savedfilename3 = input("ten topo")
savedfilename4 = input("số request")
savedfilename = savedfilename1 + str(savedfilename2) + "sw" + savedfilename3 + savedfilename4 + "Request.csv"
print(savedfilename)
df.to_csv(savedfilename, sep='\t')
df
```
